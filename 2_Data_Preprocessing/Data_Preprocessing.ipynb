## Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../student_exam_scores.csv")
```

```python
data = data.drop(columns=['student_id'])
```

```python
for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
data.isnull().sum()
```

```python
# Fill numerical missing values with median(If there is any missing value)
for col in data.columns:
    data[col] = data[col].fillna(data[col].median())
```

```python
data.to_csv("../cleaned_student_data.csv", index=False)
```
