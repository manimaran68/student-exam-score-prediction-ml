## Univariate and Bivariate

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../cleaned_student_data.csv")
```

```python
data.hist(figsize=(12,8), bins=20, edgecolor='black')
plt.suptitle("Univariate Analysis - Numerical Variables")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.title("Boxplots for Outlier Detection")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
features = ['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores']

for col in features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=data[col], y=data['exam_score'])
    plt.title(f"{col} vs exam_score")
    plt.show()
```

```python
data.to_csv("../final_student_data.csv", index=False)
```
